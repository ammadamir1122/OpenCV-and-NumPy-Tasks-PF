```python
import cv2
import numpy as np
from PIL import Image

im_path = "/content/baboon.png"

im_array = cv2.imread(im_path)

display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))

lower_red = np.array([0, 0, 150], dtype=np.uint8)
upper_red = np.array([255, 150, 255], dtype=np.uint8)


mask = cv2.inRange(im_array, lower_red, upper_red)

im_array[mask == 255] = [0, 0, 0]  # Replacing the red pixels with black

display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))  #Displaying Modified Image

```
